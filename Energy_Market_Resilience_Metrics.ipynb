Import the required modules


```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Import the required datasets

```python
hist_energy_data= pd.read_csv('historical_energy_data.csv')

hist_energy_data.head()
```

```python
infrastructure_data= pd.read_csv('infrastructure_data.csv')
print(infrastructure_data.head())
print(infrastructure_data.info())
```

```python
market_data= pd.read_csv('market_data.csv')
print(market_data.head())
print(market_data.info())
```

```python
reg_data= pd.read_csv('regulatory_data.csv')
print(reg_data.head())
print(reg_data.info())
```

```python
# Change the date columns to the same format
old_format = '%d/%m/%Y'
# First, convert the dates to datetime without specifying the format
hist_energy_data['Date/Time'] = pd.to_datetime(hist_energy_data['Date/Time'], format= old_format)
# Then, convert them to the desired format
new_format= '%Y-%m-%d'
hist_energy_data['Date/Time'] = pd.to_datetime(hist_energy_data['Date/Time'], format= new_format )
print(hist_energy_data.head())
print(hist_energy_data.info())
```

```python
# Change the date columns to the same format
old_format = '%d/%m/%Y'
# First, convert the dates to datetime without specifying the format
infrastructure_data['Date/Time'] = pd.to_datetime(infrastructure_data['Date/Time'], format= old_format)
# Then, convert them to the desired format
new_format= '%Y-%m-%d'
infrastructure_data['Date/Time'] = pd.to_datetime(infrastructure_data['Date/Time'], format= new_format )
print(infrastructure_data.head())
print(infrastructure_data.info())
```

```python
new_format = '%Y-%m-%d'
# First, convert the dates to datetime with specifying the format
market_data['Date/Time'] = pd.to_datetime(market_data['Date/Time'], format= new_format )
print(market_data.head())
print(market_data.info())
```

```python
new_format = '%Y-%m-%d'
# First, convert the dates to datetime with specifying the format
reg_data['Date/Time'] = pd.to_datetime(reg_data['Date/Time'], format= new_format )
print(reg_data.head())
print(reg_data.info())
```

```python
# For Exploratory Data Analysis
# Location/Region and Energy source from hist_energy_data
# Infrastructure status, Maintenance activities and Technological limitations from infrastructure_data
# Competitor data and market trends from market_data
# Regulatory changes and compliance status from reg_data
```

```python
# Group all datasets into a dictionary
datasets = {'infrastructure_data': infrastructure_data, 'market_data': market_data, 'regulatory_data': reg_data, 'historical_energy_data': hist_energy_data}
```

```python
# Extracting unique values and their counts for categorical columns

categorical_columns = {'infrastructure_data': ['Infrastructure Status', 'Maintenance Activities', 'Technology Limitations'], 'market_data': ['Competitor Data', 'Market Trends'], 'regulatory_data': ['Regulatory Changes', 'Compliance Status'],'historical_energy_data': ['Location/Region', 'Energy Source']}

categorical_values_counts ={}
for dataset, columns in categorical_columns.items():
    for column in columns:
        categorical_values_counts[(dataset, column)] = datasets[dataset][column].value_counts()
```

```python
# Visualizing
# Calculate the number of rows and columns for the subplots

num_rows = (len(categorical_values_counts) + 1) // 2
num_cols = 2

# Create a grid of the subplots
fig, axes = plt.subplots(num_rows, num_cols, figsize=(12, 12))

# Flatten the axes array for easier indexing
axes = axes.flatten()

# Iterate through categorical columns and plot them in subplots
for i, ((dataset, column), counts) in enumerate(categorical_values_counts.items()):
    ax = axes[i]
    sns.barplot(x=counts.values, y=counts.index, ax=ax)
    ax.set_title(f"Value Counts for {dataset} - {column}")
    ax.set_xlabel('Count')
    ax.set_ylabel(column)
    
# Hide any remaining empty subplots
for i in range(len(categorical_values_counts), num_rows * num_cols):
    fig.delaxes(axes[i])
    
# Adjust
plt.tight_layout()

plt.show()
```

**Energy demand, Production and Consumption Over Time**

We will start by examining the trends in energy demand, production and consumption from the historical_energy_dataset
This will give us insights into flunctuations and potential vulnerabilities in the energy market


```python
# Extract year and month from the date/time column
hist_energy_data['Year'] = hist_energy_data['Date/Time'].dt.year
hist_energy_data['Month'] = hist_energy_data['Date/Time'].dt.month

# Create a new column 'Year-Month' for easy plotting
hist_energy_data['Year-Month'] = hist_energy_data['Date/Time'].dt.to_period('M')

# Ensure all columns to be aggregated are numeric
numeric_columns = hist_energy_data.select_dtypes(include=['number']).columns

# Aggregating data on a monthly basis
monthly_data = hist_energy_data.groupby('Year-Month')[numeric_columns].mean()

# Setting figsize
plt.figure(figsize=(15, 6))
```

```python
# Plotting
sns.lineplot(data= monthly_data, x= monthly_data.index.astype(str), y= 'Energy Production (kWh)', label= 'Energy Production', color= 'blue', linestyle= '-', linewidth= 1.5, ci= None )

sns.lineplot(data= monthly_data, x= monthly_data.index.astype(str), y= 'Energy Consumption (kWh)', label= 'Energy Consumption', color= 'red', linestyle= '-', linewidth= 1.5, ci= None )

sns.lineplot(data= monthly_data, x= monthly_data.index.astype(str), y= 'Energy Demand', label= 'Energy Demand', color= 'green', linestyle= '-', linewidth= 1.5, ci= None )

# Setting the title and labels
plt.title('Monthly aggregate of Energy Production Consumption, and Demand Over Time')
plt.xlabel('Date')
plt.ylabel('kWh')
plt.legend(loc= 'upper left', bbox_to_anchor= (1,1))

# Setting our labels
labels= monthly_data.index.astype(str).tolist()
n= 6
plt.xticks(labels[::n], rotation= 360)

plt.tight_layout()
plt.subplots_adjust(hspace=0.5)

plt.show()
```

Looking at the plot, energy production and energy consumption seem to be almost the same throughout the months and that is expected, since, energy production do not meet energy demand. This can lead to vulnerability and a considerable loss of customers to competitors.

**Market Price and Energy Price Over Time**

To understand pricing dynamics and its relation to market trends
Visualization of thr market price from the market data dataset alongside the energy price from the historical_energy_data dataset over time

```python
# Merging historical_energy_data and market_data on Date/Time for combined analysis
merged_data= pd.merge(hist_energy_data, market_data, on= 'Date/Time')

# Setting Date/Time as the index of the DataFrame for aggregation
merged_data['Year']= merged_data['Date/Time'].dt.year
merged_data['Month']= merged_data['Date/Time'].dt.month

# Aggregating date on a monthly basis
# Ensure all columns to be aggregated are numeric
numeric_columns = merged_data.select_dtypes(include=['number']).columns

# Aggregating data on a monthly basis
monthly_merged_data = merged_data.groupby(['Year', 'Month'])[numeric_columns].mean()

# Create a new 'Year-Month' column for plotting
monthly_merged_data['Year-Month']= monthly_merged_data.index.get_level_values(0).astype(str) + '-' + monthly_merged_data.index.get_level_values(1).astype(str)

plt.figure(figsize= (15, 6))

sns.lineplot(data= monthly_merged_data, x= 'Year-Month', y= 'Market Price', label= 'Market Price', color= 'green', linestyle= '-', linewidth= 1.5, ci= None )

sns.lineplot(data= monthly_merged_data, x= 'Year-Month', y= 'Energy Price', label= 'Energy Price', color= 'yellow', linestyle= '-', linewidth= 1.5, ci= None )

plt.title('Monthly Aggregate of Market Price and Energy Price Over Time')

plt.xlabel('Date')
plt.ylabel('Price ($)')
plt.legend(loc= 'upper left', bbox_to_anchor= (1,1))

# Setting our labels
plt.grid(True, which= 'both', linestyle= '-', linewidth= 0.5)
labels= monthly_merged_data['Year-Month'].tolist()
n= 6
plt.xticks(labels[::n], rotation= 360)
plt.tight_layout()

plt.show()
```

**Energy Demand and Energy Price correlation**
Let's check if the energy demand is influenced by the current energy price

```python
correlation_energy_price= merged_data['Energy Demand'].corr(merged_data['Energy Price'])

correlation_market_price= merged_data['Market Demand'].corr(merged_data['Market Price'])

print(correlation_energy_price)

print(correlation_market_price)
```

**Infrastructure status and Technology limitation**

To understand the state of infrastructure and Technology over time

Visualize the frequency of various infrastructure values and the distribution of Technology limitations from infrastructure_data dataset

```python
# Setting up the figure and axes
fig, axes = plt.subplots(1, 2, figsize=(15, 8))

#P Plotting the frequency of Infrastructure status
sns.countplot(data=infrastructure_data, x='Infrastructure Status', ax=axes[0], order=['Good', 'Fair', 'Poor'], palette='viridis')
axes[0].set_title('Frequency of Infrastructure status')
axes[0].set_xlabel('Infrastructure status')
axes[0].set_ylabel('count')

# Plotting the frequency of Technology limitation
sns.countplot(data=infrastructure_data, x='Technology Limitations', ax=axes[1], palette='viridis')
axes[1].set_title('Frequency of Technology Limitation')
axes[1].set_xlabel('Technology Limitations')
axes[1].set_ylabel('count')

plt.tight_layout()
plt.show()
```

**Check the relationship between infrastructure status and technology limitation**

```python
ct = pd.crosstab(infrastructure_data['Infrastructure Status'], infrastructure_data['Technology Limitations'])

plt.figure(figsize=(10, 6))

sns.heatmap(ct, annot=True, cmap='viridis')
plt.title('Infrastructure status vs. Technology limitations')
plt.xlabel('Technology Limitations')
plt.ylabel('Infrastructure Status')

plt.tight_layout()
plt.show()
```

**Correlation between poor infrastructure status and high technology limitation and the energy production**

```python
infrastructure_data['Poor_Infrastructure'] = infrastructure_data['Infrastructure Status'].apply(lambda x: 1 if x == 'Poor' else 0)
infrastructure_data['High_Tech_limitations'] = infrastructure_data['Technology Limitations'].apply(lambda x: 1 if x == 'High' else 0)

infrastructure_data['combined'] = infrastructure_data['Poor_Infrastructure'] * infrastructure_data['High_Tech_limitations']

merged_data = pd.merge(hist_energy_data, infrastructure_data, on='Date/Time')

correlation_coefficient = merged_data['Energy Production (kWh)'].corr(merged_data['combined'])

print(correlation_coefficient)
```

**Regulatory Changes and Compliance Costs**

To understand the impact of Regulatory Changes

Visualize the frequency of various Regulatory Changes and the distribution of Compliance Costs associated with these changes from the regulatory dataset

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

sns.countplot(data=reg_data, x='Regulatory Changes', ax=ax1, palette='viridis')
ax1.set_title('Frequency of Regulatory Changes')
ax1.set_xlabel('Regulatory Changes')
ax1.set_ylabel('Count')

sns.histplot(data=reg_data, x='Compliance Costs', ax=ax2, bins=30, kde=True, color='purple')
ax2.set_title('Frequency of Compliance Costs')
ax2.set_xlabel('Compliance Costs')
ax2.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

**Visualize the relationship between regulatory changes and the associated compliance costs as well as the operational cost to the revenue generated by the company**

Revenue= Energy consumption * Energy price

This will provide insights into the financial implications of regulatory changes and operational cost for Energix Enterprise

```python
hist_energy_data['Revenue'] = hist_energy_data['Energy Price'] * hist_energy_data['Energy Consumption (kWh)']

merged_reg_data = pd.merge(hist_energy_data, reg_data, on='Date/Time')

# Ensure 'Date/Time' column is in datetime format
merged_reg_data['Date/Time'] = pd.to_datetime(merged_reg_data['Date/Time'])

# Extract Year and Month from 'Date/Time' column
# Use a different column name to avoid conflict
merged_reg_data['Year'] = merged_reg_data['Date/Time'].dt.year
merged_reg_data['Month_Extracted'] = merged_reg_data['Date/Time'].dt.month

# Ensure that only numeric columns are aggregated
numeric_columns = merged_reg_data.select_dtypes(include=['number']).columns
print('Numeric columns: ', numeric_columns)
print('Value: ',[numeric_columns])
monthly_aggregated_data = merged_reg_data.groupby(['Year', 'Month_Extracted'])[numeric_columns].mean().reset_index(drop=True)

monthly_aggregated_data['Year-Month'] = monthly_aggregated_data['Year'].astype(str) + '-' + monthly_aggregated_data['Month_Extracted'].astype(str)

plt.figure(figsize=(15, 6))

plt.fill_between(monthly_aggregated_data['Year-Month'], monthly_aggregated_data['Operational Costs'], color='blue', label='Operational Costs', alpha=0.5)
plt.fill_between(monthly_aggregated_data['Year-Month'], monthly_aggregated_data['Operational Costs'], monthly_aggregated_data['Operational Costs'] + monthly_aggregated_data['Compliance Costs'], color='green', label='Compliance Costs', alpha=0.5)

sns.lineplot(monthly_aggregated_data, x='Year-Month', y='Revenue', label='Revenue', color='red', linewidth=2)

plt.title('Impact of Regulatory changes on Operational, Revenue, and Compliance Costs (Monthly Aggregated)')
plt.xlabel('Date')
plt.ylabel('Amount ($)')
plt.legend(loc='upper left')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
labels= monthly_aggregated_data['Year-Month'].tolist()
n=4
plt.xticks(labels[::n], rotation=360)

plt.tight_layout()

plt.show()
```

**Analyzing Competitions from Renewable Energy Providers**

To visualize the competition from renewable energy providers and its impact on Energix Enterprise, analyze the Energy Source column from the hist_energy_data dataset

Visualize the trends in energy production based on the energy source (Fossil Fuels vs. Renewable) over time. This will give insight into how the productions affected the company's market share

```python
hist_energy_data['Year'] = hist_energy_data['Date/Time'].dt.year
hist_energy_data['Month'] = hist_energy_data['Date/Time'].dt.month

# Ensure only numeric columns are summed
monthly_aggregated_data = hist_energy_data.groupby(['Year', 'Month', 'Energy Source'])['Energy Production (kWh)'].sum().reset_index()

monthly_aggregated_data['Year-Month'] = monthly_aggregated_data['Year'].astype(str) + '-' + monthly_aggregated_data['Month'].astype(str)

plt.figure(figsize=(12, 4))

sns.lineplot(data=monthly_aggregated_data, x='Year-Month', y='Energy Production (kWh)', hue='Energy Source', ci=None)


plt.title('Monthly Aggregated Energy Production Trends by Energy Source Over Time')
plt.xlabel('Date')
plt.ylabel('Energy Production (kWh)')
plt.legend(title='Energy Source')
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
labels= monthly_aggregated_data['Year-Month'].tolist()
n=4
plt.xticks(labels[::n], rotation=360)

plt.tight_layout()

plt.show()
```
